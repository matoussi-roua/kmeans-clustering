K-Means clustering is used to find intrinsic groups within the unlabelled dataset and draw inferences from them. It is based on centroid-based clustering.

K-Means clustering is the most common unsupervised machine learning algorithm. 
It is widely used for many applications which include:
Image segmentation
Customer segmentation
Species clustering
Anomaly detection
Clustering languages

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
x,y=make_blobs(n_samples=100,centers=5,cluster_std=0.9) #construction dataset # cluster_std : distance entre les points
```

```python
x  #x1 et x2
```

```python
y
```

```python
plt.scatter(x[:,0],x[:,1])
```

#elbow method : determiner le nombre exacte de clusters
The elbow method is used to determine the optimal number of clusters in K-means clustering. The elbow method plots the value of the cost function produced by different values of K. 

```python
inertia=[]
k_range=range(1,20)
for k in k_range:
    model=KMeans(n_clusters=k).fit(x)
    # ou bien
    #model=KMeans(n_clusters=k)  : nb of clusters
    #model.fit(x)  :execute
    inertia.append(model.inertia_) #inertia c est une liste des couts differents classes par ordre de k
inertia
        
```

```python
plt.plot(k_range,inertia)  #on choisit le cout minimal et on lit k
plt.scatter(5,0,c='r')
```

creation of KMeans model :

```python
model=KMeans(n_clusters=5) #
```

```python
model.fit(x) #execution du model
```

```python
model.labels_
```

```python
model.predict(x) #testing data
```

```python
plt.scatter(x[:,0],x[:,1],c=model.predict(x))
```

```python
vect=model.cluster_centers_ #positions des clusters
vect
```

```python
plt.scatter(vect[:,0],vect[:,1],c='r')
```

```python
plt.scatter(x[:,0],x[:,1],c=model.predict(x))
plt.scatter(vect[:,0],vect[:,1],c='r')
```

```python
model.inertia_
```

```python
model.score(x) #le cout negatif
```
